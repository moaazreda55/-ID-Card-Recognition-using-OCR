```python
import cv2
import matplotlib.pyplot as plt
import pytesseract
import pandas as pd
import numpy as np
from PIL import Image
```

```python
img=cv2.imread('D:/sql/Data set/images/id-3.PNG') 
```

```python
plt.imshow(img)
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
blurred = cv2.GaussianBlur(gray, (5, 5), 0)
thresh = cv2.threshold(blurred, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)[1]
```

```python
pytesseract.pytesseract.tesseract_cmd = r"C:\Users\HP\AppData\Local\Tesseract-OCR\tesseract.exe"
```

```python
text = pytesseract.image_to_string(thresh, config="--psm 4")
```

```python
text.splitlines()
```

```python
columns_lst = [["Country", "Name", "ID", "Gender"],
               [text.splitlines()[0], text.splitlines()[2], text.splitlines()[7], text.splitlines()[21]]]

df = pd.DataFrame(columns_lst, columns=["Country", "Name", "ID", "Gender"])
df 
```
